```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

import cv2 as cv
```

```python
def show_in_gray(image_array):
    image_array = image_array[:, :, 0]  
    plt.imshow(image_array, cmap="gray")
    plt.colorbar()
    plt.show()
    
```

```python
#_, img_bin = cv.threshold(img, T, 255, cv2.THRESH_BINARY)
_, img_otsu = cv.threshold(Iafter, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(img_otsu, cmap="gray")
```

```python
img_adapt = cv.adaptiveThreshold(Iafter, 255, cv.ADAPTIVE_THRESH_MEAN_C,
                                  cv.THRESH_BINARY, blockSize=11, C=2)
plt.imshow(img_adapt, cmap="gray")
```

```python
Ihistdyn = cv.imread("ressources/Ihistdyn.png",cv.IMREAD_GRAYSCALE)
a = np.max(Ihistdyn)
b = np.min(Ihistdyn)

def dynamic_change(img, a, b):
    img_stretch = (img - b) / (a - b) * 255
    return np.clip(img_stretch, 0, 255).astype(np.uint8)

Ihistegal = dynamic_change(Ihistdyn,a,b)
plt.imshow(Ihistdyn,cmap="gray")
```
